3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
df['Stress Levels'].describe()
````

Уровень стресса распределен от 1 до 10, причем он представлен в виде float и в таком виде с ним сложно работать. Разобьем его на 3 группы: "low" для низкого уровня (менее 3), "high" для высокого (больше 8), остальное "medium" 

````python
df_task_3 = df.loc[:,['Stress Levels','Blood Pressure (s/d)']] # выделяю отдельную выборку из 2-х столбцов

# Делю столбец стресса на 3 группы
df_task_3['Stress Levels'] = df_task_3['Stress Levels'].apply(lambda x: "low" if x < 3 else "high" if x > 8 else "medium") 

# Разделяю столбец кровяного давления на 2
df_task_3['Systolic'] = df_task_3['Blood Pressure (s/d)'].apply(lambda x: x.split('/')[0]).astype('int')
df_task_3['Diastolic'] = df_task_3['Blood Pressure (s/d)'].apply(lambda x: x.split('/')[1]).astype('int')

# Удаляю лишний столбец
df_task_3 = df_task_3.drop(['Blood Pressure (s/d)'],axis = 1)

# Группирую по уровню стресса
df_task_3.groupby(['Stress Levels'])[['Systolic','Diastolic']].agg(["mean", "std", "min", "max"])
````

Как видно значения для высокого уровня стресса и для низкого примерно одинаковые
**Ответ:Нет**

5. Постройте круговую диаграмму отображающую как ментальное здоровье связано с паттернами сна

````python
# Делю данные на 4 выборки по состоянию ментального здоровья
df_mental_good = df[df['Mental Health Status'] == 'Good']
df_mental_poor = df[df['Mental Health Status'] == 'Poor']
df_mental_fair = df[df['Mental Health Status'] == 'Fair']
df_mental_excellent = df[df['Mental Health Status'] == 'Excellent']

labels = list(df['Sleep Patterns'].unique()) # список паттернов сна


data_good = [len(df_mental_good[df_mental_good['Sleep Patterns'] == 'Insomnia']),len(df_mental_good[df_mental_good['Sleep Patterns'] == 'Normal']),\
             len(df_mental_good[df_mental_good['Sleep Patterns'] == 'Excessive'])]

data_poor = [len(df_mental_poor[df_mental_poor['Sleep Patterns'] == 'Insomnia']),len(df_mental_poor[df_mental_poor['Sleep Patterns'] == 'Normal']),\
             len(df_mental_poor[df_mental_poor['Sleep Patterns'] == 'Excessive'])]

data_fair = [len(df_mental_fair[df_mental_fair['Sleep Patterns'] == 'Insomnia']),len(df_mental_fair[df_mental_fair['Sleep Patterns'] == 'Normal']),\
             len(df_mental_fair[df_mental_fair['Sleep Patterns'] == 'Excessive'])]

data_excellent = [len(df_mental_excellent[df_mental_excellent['Sleep Patterns'] == 'Insomnia']),\
                  len(df_mental_excellent[df_mental_excellent['Sleep Patterns'] == 'Normal']),\
                  len(df_mental_excellent[df_mental_excellent['Sleep Patterns'] == 'Excessive'])]

# Настройка цветовой гаммы
colors = sns.color_palette('pastel')[0:5]

# Размер отрисовки
plt.figure(figsize=[11, 9])

# Общий заголовок
plt.suptitle('Зависимость ментального здоровья с паттернами сна')

# График 1
plt.subplot(2, 2, 1)
plt.pie(data_good, labels = labels, colors = colors, autopct='%.0f%%')
plt.title("Mental Health Status: Good")

# График 2
plt.subplot(2, 2, 2)
plt.pie(data_poor, labels = labels, colors = colors, autopct='%.0f%%')
plt.title("Mental Health Status: Poor")

# График 3
plt.subplot(2, 2, 3)
plt.pie(data_fair, labels = labels, colors = colors, autopct='%.0f%%')
plt.title("Mental Health Status: Fair")

# График 4
plt.subplot(2, 2, 4)
plt.pie(data_excellent, labels = labels, colors = colors, autopct='%.0f%%')
plt.title("Mental Health Status: Excellent")

plt.show()
````
